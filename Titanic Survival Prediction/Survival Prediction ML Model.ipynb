# Titanic Survival Prediction (Logistic Regression)

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
```

```python
tr_df = pd.read_csv("Cleaned_Train_Data.csv")
```

```python
encoded_tr = pd.get_dummies(tr_df, columns = ['Pclass', 'Sex', "Embarked"], drop_first=True) 
print(encoded_tr)
```

```python
X= encoded_tr.iloc[:,2:].values
y= encoded_tr.iloc[:,1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
classifier = LogisticRegression()
```

```python
params = {"penalty":['l1', 'l2', 'elasticnet'], "max_iter":[10, 20, 30, 40, 50, 75, 100, 200, 250, 300], "C":[1,2,3,4,5,10,20,30,40,50]}
```

```python
classifier_rgsr = GridSearchCV(classifier, param_grid=params,scoring="accuracy",cv=5)
```

```python
classifier_rgsr.fit(X_train, y_train)
```

```python
print(classifier_rgsr.best_params_)
```

```python
y_pred = classifier_rgsr.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

A Logistic Regression model is developed with about 81% accuracy.
